```python
def atom_feature(atom):
    return np.array(one_of_k_encoding_unk(atom.GetSymbol(),
                                      ['C', 'N', 'O', 'S', 'F', 'H', 'Si', 'P', 'Cl', 'Br',
                                       'Li', 'Na', 'K', 'Mg', 'Ca', 'Fe', 'As', 'Al', 'I', 'B',
                                       'V', 'Tl', 'Sb', 'Sn', 'Ag', 'Pd', 'Co', 'Se', 'Ti', 'Zn',
                                       'Ge', 'Cu', 'Au', 'Ni', 'Cd', 'Mn', 'Cr', 'Pt', 'Hg', 'Pb']) +
                    one_of_k_encoding(atom.GetDegree(), [0, 1, 2, 3, 4, 5]) +
                    one_of_k_encoding_unk(atom.GetTotalNumHs(), [0, 1, 2, 3, 4]) +
                    one_of_k_encoding_unk(atom.GetImplicitValence(), [0, 1, 2, 3, 4, 5]) +
                    [atom.GetIsAromatic()])    # (40, 6, 5, 6, 1)
```

```python
def one_of_k_encoding(x, allowable_set):
    if x not in allowable_set:
        raise Exception("input {0} not in allowable set{1}:".format(x, allowable_set))
    #print list((map(lambda s: x == s, allowable_set)))
    return list(map(lambda s: x == s, allowable_set))

def one_of_k_encoding_unk(x, allowable_set):
    """Maps inputs not in the allowable set to the last element."""
    if x not in allowable_set:
        x = allowable_set[-1]
    return list(map(lambda s: x == s, allowable_set))
```

```python
def convert_to_graph(InChi_list):
    adj = []
    adj_norm = []
    features = []
    maxNumAtoms = 50
    for i in InChi_list:
        #Mol
        iMol = Chem.inchi.MolFromInchi(i.strip())
        iMol = Chem.rdmolops.AddHs(iMol)
        #Adj
        iAdjTmp = Chem.rdmolops.GetAdjacencyMatrix(iMol)
        iMol = mol_with_atom_index(iMol)
        # Feature
        if(iAdjTmp.shape[0] <= maxNumAtoms):
            # Feature-preprocessing
            iFeature = np.zeros((maxNumAtoms, 58))
            iFeatureTmp = []
            for atom in iMol.GetAtoms():
                arr = atom_feature(atom)
                iFeatureTmp.append(arr)
            iFeature[0:len(iFeatureTmp), 0:58] = iFeatureTmp ### 0 padding for feature-set
            features.append(iFeature)

            # Adj-preprocessing
            iAdj = np.zeros((maxNumAtoms, maxNumAtoms))
            iAdj[0:len(iFeatureTmp), 0:len(iFeatureTmp)] = iAdjTmp + np.eye(len(iFeatureTmp))
            adj.append(np.asarray(iAdj))
    features = np.asarray(features)

    return features, adj
    
```

# 3. Model Architecture

```python
class inv_wcal(nn.Module):
    def __init__(self, in_dim, out_dim):
        super(inv_wcal, self).__init__()
        self.in_dim = in_dim
        self.out_dim = out_dim
        self.conv = Conv(self.in_dim, self.out_dim)
    
    def forward(self, input_X):
        inverse_X = self.inverse(input_X)
        _X = self.conv(inverse_X)
        output_X = self.inverse(_X)
        
        return output_X
    
    def inverse(self, input_X):
        batch_size, num_atom, num_featue = input_X.size()
        s = list()
        for i in range(batch_size):
            x = torch.t(torch.squeeze(torch.chunk(input_X, batch_size, dim = 0)[i]))
            s.append(x)
        inverse_X = torch.stack(s, dim = 0)
        
        return inverse_X

class Conv(nn.Module):
    def __init__(self, in_dim, out_dim):
        super(Conv, self).__init__()
        self.in_dim = in_dim
        self.out_dim = out_dim
        self.conv = nn.Conv1d(in_dim, out_dim, 1)
    
    def forward(self, input_X):
        output_X = self.conv(input_X)
        
        return output_X
```

 ## A. Skip Conneciton
We already have seen this improvements for vanilla model in the CNN assignment. The skip connection model's purpose is that preventing the gradient vanishing problem in the deep network model, by adding the input data of each operation into the ouput data of each operations. This can be expressed with the below formula.
\begin{equation*}
H^{l+1}= H^l+ H^{l-1} 
\end{equation*}

This method has improved version which is called gated-skip connection model. Gated means that we gives a weight to the tendency of getting features from the input data. This can be expressed with the below formula.
\begin{equation*}
H^{l+1}= z H^l+ (1-z) H^{l-1} 
\end{equation*}
In the formula, the coefficient z is called 'gate coefficient' and calculated by applying sigmoid function to sum of input, output data.

```python
class skip_connection(nn.Module):
    def __init__(self, in_dim, out_dim):
        super(skip_connection, self).__init__()
        self.in_dim = in_dim
        self.out_dim = out_dim
        self.weight = inv_wcal(self.in_dim, self.out_dim)
    
    def forward(self, input_X, new_X):
        _input_X = self.weight(input_X) if self.in_dim != self.out_dim else input_X
        
        output_X = _input_X + new_X
        
        return output_X
    
class gated_skip_connection(nn.Module):
    def __init__(self, in_dim, out_dim):
        super(gated_skip_connection, self).__init__()
        self.in_dim = in_dim
        self.out_dim = out_dim
        self.weight_in = inv_wcal(self.in_dim, self.out_dim)
        #self.weight_new = inv_wcal(self.out_dim, self.out_dim)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, input_X, new_X):
        z = self.gate_coefficient(input_X, new_X)
        _input_X = self.weight_in(input_X) if self.in_dim != self.out_dim else input_X
        
        output_X = torch.mul(new_X, z) + torch.mul(_input_X, 1.0-z)
        
        return output_X
    
    def gate_coefficient(self, input_X, new_X):
        X1 = self.weight_in(input_X)
        #X2 = self.weight_new(new_X)
        X2 = new_X
        
        gate_coefficient = self.sigmoid(X1 + X2)
        
        return gate_coefficient
```

## B. Readout and Attention
Readout is necessary method for graph convolution network model. This allows us to get the same results from the different permutations of graph nodes(permutation invariance). Summation over all existing nodes in the
graph satisfies the permutation invariance.
Attention is another way to improve the vanilla GCN model. This is some similarity with the gated-skip connection because this gives a weight to tendency of each nodes influencing another nodes. This can be implemented by a attention matrix, which can be calculated with the below formula.
\begin{equation*}
e_{i,j} = \sigma({C (H_i W) ({H_j W}^T)})
\end{equation*}
In this formula, ith components of H means ith atoms' featues(vector size of 1 x N), and W is weight matrix which allows learning of the attention matrix(vector size of N x N).

```python
class Readout(nn.Module):
    def __init__(self, in_dim, out_dim, activation):
        super(Readout, self).__init__()
        self.in_dim = in_dim
        self.out_dim = out_dim
        self.activation = activation
        self.weight = inv_wcal(self.in_dim, self.out_dim)
        
    def forward(self, input_X):
        output_X1 = self.weight(input_X)
        output_X2 = output_X1.sum(1)
        output_X = self.activation(output_X2)
        
        return output_X

def new_parameter(*size):
    out = nn.Parameter(torch.FloatTensor(*size))
    torch.nn.init.xavier_normal_(out)
    return out

class Attention(nn.Module):
    def __init__(self, attn_size):
        super(Attention, self).__init__()
        self.attn_size = attn_size
        self.attn = new_parameter(self.attn_size, 1)
        self.relu = nn.ReLU()
        self.softmax = nn.Softmax(dim = 0)
        
    def forward(self, input_X):
        output_X = self.relu(self.attn_cal(input_X))
        
        return output_X
        
    def attn_cal(self, X):
        batch_size, num_atom, num_feature = X.size()
        # num_feature = attn_size
        # (batch_size * num_atom * num_feature) (attn_size * 1) -> batch_size * num_atom * 1 -> batch_size * num_atom
        # -> softmax -> batch_size * num_atom * 1 -> batch_size * num_atom * num_feature
        attn_matrix = (torch.matmul(X, self.attn)).squeeze() 
        attn_matrix = self.softmax(attn_matrix)
        attn_matrix = attn_matrix.view(X.size(0), X.size(1), 1)
        attn_matrix = X * attn_matrix
        
        return attn_matrix
```

## C. Graph Convolutional Network model
There are three big parts in my GCN model.
### I. GCN block
GCN block can be expressed by the below formula.
\begin{equation*}
H^{l+1} = A(H^l W^l + B^l)
\end{equation*}
My GCN block has attention method and can apply gated-skip connection, skip connection or no skip connection. Also, if I previously setted a number of layers then then number ofthe blocks between second block to the block before readout should be number of layers - 1.
### II. Readout
Readout is applied to the products of the n-layer GCN blocks. So that it makes the result permutation invariance.
### III. Prediction layer
Prediction layer is same as before. It's just composed with multi-layer perceptron. The final output value has dimension 1, which is same dimension as vector of logP values. With the predicted logP value and logP value, we can calculate the MSE, loss, MAE, Std values.

```python
class GCNLayer(nn.Module):
    def __init__(self, in_dim, out_dim, attn):
        super(GCNLayer, self).__init__()
        self.in_dim = in_dim
        self.out_dim = out_dim
        self.attn = attn #True or False
        self.weight = inv_wcal(self.in_dim, self.out_dim)
        self.attention = Attention(self.out_dim)

    # graph_convolution = H^(l+1) = A{H^(l)W^(l)+b^(l))
    # size of input_X = batch_size x num_atoms x num_features
    # size of input_A = batch_size x num_atoms x num_atoms
    def forward(self, input_X, input_A):
        if not self.attn: # if False
            output_X = torch.mm(input_A, self.weight(input_X))
            return output_X
        else:
            output_X = self.weight(input_X)
            output_X = torch.mm(input_A, self.attention(output_X))
            return output_X

class GCNBlock(nn.Module):
    def __init__(self, n_atom, in_dim, out_dim, attn, sc, bn, dp_rate):
        super(GCNBlock, self).__init__()
        self.n_atom = n_atom
        self.in_dim = in_dim
        self.out_dim = out_dim
        self.attn = attn
        self.sc = sc
        self.bn = bn
        self.dropout = nn.Dropout(dp_rate)
        self.layer = GCNLayer(self.in_dim, self.out_dim, self.attn)
        
    def fowrard(self, input_X, input_A):
        if self.bn:
            m = nn.Batchnorm1d(self.n_atom)
            input_X = m(input_X)
            
        _X = input_X
        output_X = GCNLayer(input_X, input_A)
        
        if self.sc == 'no':
            output_X = self.dropout(output_X)
            
        elif self.sc == 'sc':
            self.skip_connection = skip_connection(self.in_dim, self.out_dim)
            output_X = self.dropout(self.skip_connection(output_X, _X))
            
        elif self.sc == 'gsc':
            self.skip_connection = gated_skip_connection(self.in_dim, self.out_dim)
            output_X = self.dropout(self.skip_connection(output_X, _X))
        
        return output_X

class predictor(nn.Module):
    def __init__(self, in_dim, out_dim, activation = None):
        super(predictor, self).__init__()
        self.in_dim = in_dim
        self.out_dim = out_dim
        self.linear = nn.Linear(self.in_dim, self.out_dim)
        nn.init.xavier_uniform_(self.linear.weight)
        self.activation = activation
        
    def forward(self, input_X):
        output_X = self.linear(input_X)
        output_X = self.activation(output_X)
        
        return output_X

class NN(nn.Module):
    def __init__(self, args):
        super(NN, self).__init__()
        # args.n_layer, args.in_dim, args.hidded_dim, args.pred_dim1, args.pred_dim2, args.pred_dim3
        # args.out_dim, args.dp_rate, args.sc, args.attn, args.n_atom
        '''
        self.n_layer = args.n_layer
        self.gcn1 = GCN(args.in_dim, args.hidden_dim, args.drop_rate, args.n_atom,nn.ReLU(), args.sc, args.attention)
        self.gcn2 = GCN(args.hidden_dim, args.hidden_dim, args.drop_rate, args.n_atom,nn.ReLU(), args.sc, args.attention)
        '''
        self.readout = Readout(args.hidden_dim, args.pred_dim1, nn.ReLU())
        self.ppm_predictor1 = predictor(args.pred_dim1, args.pred_dim2, args.activation)
        self.ppm_predictor2 = predictor(args.pred_dim2, args.pred_dim3, args.activation)
        self.ppm_predictor3 = nn.Linear(args.pred_dim3, args.out_dim)
        self.int_predictor1 = predictor(args.pred_dim1, args.pred_dim2, args.activation)
        self.int_predictor2 = predictor(args.pred_dim2, args.pred_dim3, args.activation)
        self.int_predictor3 = nn.Linear(args.pred_dim3, args.out_dim)
        
        GCN = nn.ModuleList()
        GCN.append(GCNBlock(args.n_atom, args.in_dim, args.hidden_dim, args.attn, args.sc, args.bn, args.dp_rate))
        GCN.extend([GCNBlock(args.n_atom, args.hidden_dim, args.hidden_dim, args.attn, args.sc, args.bn, args.dp_rate) for i in range(1, args.n_layer)])
        self.GCN = GCN
                   
    def forward(self, input_X, input_A):
        output_X = gcn(input_X, input_A)
        output_X = self.readout(output_X)
        ppm_output_X = self.ppm_predictor1(output_X)
        ppm_output_X = self.ppm_predictor2(ppm_output_X)
        ppm_output_X = self.ppm_predictor3(ppm_output_X)
        int_output_X = self.int_predictor1(output_X)
        int_output_X = self.int_predictor2(int_output_X)
        int_output_X = self.int_predictor3(int_output_X)
        
        return ppm_output_X, int_output_X
    
    def gcn(self, input_X, input_A):
        for i, module in enumerate(self.GCN):
            output_X = module(input_X, input_A)
        return output_X
```

# 4. Train, Validate, Evaluate

```python
args.batch_size = 10
args.lr = 0.001
args.l2_coef = 0.001
args.optim = 'Adam'
args.epoch = 20
args.n_layer = 4
args.in_dim = 58
args.hidden_dim = 64
args.pred_dim1 = 256
args.pred_dim2 = 128
args.pred_dim3 = 64
args.n_atom = 50
args.out_dim = 7
args.attn = True
args.dp_rate = 0.2
args.activation = nn.ReLU()
args.sc = 'gsc'

# args.n_layer, args.in_dim, args.hidded_dim, args.pred_dim1, args.pred_dim2, args.pred_dim3
# args.out_dim, args.dp_rate, args.sc, args.attn, args.n_atom

def testing(partition_dictionary, args):
    
    #args.attention = attention
            
    time_start = time.time()
    model = NN(args)

    if args.optim == 'Adam':
        optimizer = optim.Adam(model.parameters(), lr = args.lr, weight_decay = args.l2_coef)
    elif args.optim == 'RMSprop':
        optimizer = optim.RMSprop(model.parameters(), lr = args.lr, weight_decay = args.l2_coef)
    elif args.optim == 'SGD':
        optimizer = optim.SGD(model.parameters(), lr = args.lr, weight_decay = args.l2_coef)
    else:
        raise Exception("Unidentified Optimizer Type")

    criterion = nn.MSELoss()
    #att = 'With attention' if args.attention else 'Without attention'

    data_train = DataLoader(partition_dictionary['train'],
                            batch_size = args.batch_size,
                            shuffle = True)
    data_validation = DataLoader(partition_dictionary['validation'],
                                 batch_size = args.batch_size,
                                 shuffle = True)

    ppm_list_train_loss = list()
    int_list_train_loss = list()
    ppm_list_validation_loss = list()
    int_list_validation_loss = list()
    
    #attention = list()
    time_list = list()
    time_list.append(time_start)
    Epoch = list()
    
    #ppm_output_X, int_output_X
    for epoch in range(args.epoch):
        Epoch.append(epoch)
        pred_train = list()
        model.train()
        ppm_epoch_train_loss = 0
        int_epoch_train_loss = 0
        for i, batch in enumerate(data_train):
            feature_train_list = torch.tensor(batch[0])
            adj_train_list = torch.tensor(batch[1])
            ppm_train_list = torch.tensor(batch[2])
            int_train_list = torch.tensor(batch[3])
            ppm_train_list = ppm_train_list.view(-1,1)
            int_train_list = int_train_list.view(-1,1)
            #logP_train_list = logP_train_list.view(-1, 1)

              '''Since the backward() function accumulates gradients
              and you donâ€™t want to mix up gradients between minibatches
              you have to zero them out at the start of a new minibatch'''
            optimizer.zero_grad()
            list_pred_ppm, list_pred_int = model(feature_train_list, adj_train_list)
            list_pred_ppm.require_grad = False
            list_pred_int.require_grad = False
            ppm_train_loss = 0
            int_train_loss = 0
            for i in range(7):
                ppm_train_loss += criterion(ppm_train_list[i], list_pred_ppm[i])
                ppm_epoch_train_loss += ppm_train_loss.item()
                int_train_loss += criterion(int_train_list[i], list_pred_int[i])
                int_epoch_train_loss += int_train_loss.item()
            ppm_train_loss.backward()
            int_train_loss.backward()
            optimizer.step()
            
        ppm_epoch_train_loss /= len(data_train)
        int_epoch_train_loss /= len(data_train)
        ppm_list_train_loss.append(ppm_epoch_train_loss)
        int_list_train_loss.append(int_epoch_train_loss)

        pred_validation = list()
        model.eval()
        epoch_validation_loss = 0
        for i, batch in enumerate(data_validation):
            feature_validation_list = torch.tensor(batch[0])
            adj_validation_list = torch.tensor(batch[1])
            ppm_validation_list = torch.tensor(batch[2])
            int_validation_list = torch.tensor(batch[3])
            ppm_validation_list = ppm_validation_list.view(-1,1)
            int_validation_list = int_validation_list.view(-1,1)
            #logP_validation_list = logP_validation_list.view(-1, 1)

            optimizer.zero_grad()
            list_pred_ppm, list_pred_int = model(feature_train_list, adj_train_list)
            list_pred_ppm.require_grad = False
            list_pred_int.require_grad = False
            ppm_validation_loss = 0
            int_validation_loss = 0
            for i in range(7):
                ppm_validation_loss += criterion(ppm_validation_list[i], list_pred_ppm[i])
                ppm_epoch_validation_loss += ppm_validation_loss.item()
                int_validation_loss += criterion(int_validation_list[i], list_pred_int[i])
                int_epoch_validation_loss += int_validation_loss.item()
            ppm_validation_loss.backward()
            int_validation_loss.backward()
            optimizer.step()

        ppm_epoch_validation_loss /= len(data_validation)
        int_epoch_validation_loss /= len(data_validation)
        ppm_list_validation_loss.append(ppm_epoch_validation_loss)
        int_list_validation_loss.append(int_epoch_validation_loss)

        #attention.append(att)
          '''time_ = time.time()
          time_list.append(time_)
          time_epoch = time_list[epoch + 1] - time_list[epoch]
          print("{:5.3f}sec at epoch {:2}".format(time_epoch, epoch))'''
                
    data_test = DataLoader(partition_dictionary['test'],
                           batch_size = args.batch_size,
                           shuffle = args.shuffle)

    model.eval()
    #logP_total = list()
    #pred_logP_total = list()
    #attention_total = list()
    ppm_total = list()
    pred_ppm_total = list()
    int_total = list()
    pred_int_total = list()
    for i, batch in enumerate(data_test):
        feature_list = torch.tensor(batch[0])
        adj_list = torch.tensor(batch[1])
        ppm_list = torch.tensor(batch[2])
        int_list = torch.tensor(batch[3])
        ppm_total += ppm_list.tolist()
        int_total += int_list.tolist()
        ppm_list = ppm_list.view(-1, 1)
        int_list = int_list.view(-1, 1)
        #logP_total += logP_list.tolist()
        #logP_list = logP_list.view(-1, 1)
        
        list_pred_ppm, list_pred_int = model(feature_list, adj_list)
        pred_ppm_total += list_pred_ppm.view(-1).tolist()
        pred_int_total += list_pred_int.view(-1).tolist()

    #for i in range(len(logP_total)):
    #    attention_total.append(att)

    ppm_mae = mean_absolute_error(ppm_total, pred_ppm_total)
    ppm_std = np.std(np.array(ppm_total) - np.array(pred_ppm_total))
    int_mae = mean_absolute_error(int_total, pred_int_total)
    int_std = np.std(np.array(int_total) - np.array(pred_int_total))
    
    time_end = time.time()
    time_required = time_end - time_start
    '''if args.attention:
        print("With attention, got MAE: {:2.3f}, Std: {:2.3f}, Total time: {:5.3f}sec".format(mae, std, time_required))
    else:
        print("Without attention, got MAE: {:2.3f}, Std: {:2.3f}, Total time: {:5.3f}sec".format(mae, std, time_required))'''
    print("ppm MAE: {:2.3f}, ppm Std: {:2.3f}, intensity MAE: {:2.3f}, intensity Std: {:2.3f}, Total time: {:5.3f}sec".format(ppm_mae, ppm_std, int_mae, int_std, time_required))
    
    ppm_result_total = dict()
    ppm_result_total['ppm_total'] = ppm_total
    ppm_result_total['pred_ppm_total'] = pred_ppm_total
    #result_total['attention_total'] = attention_total
    int_result_total = dict()
    int_result_total['int_total'] = int_total
    int_result_total['pred_int_total'] = pred_int_total
    
    ppm_result_loss = dict()
    ppm_result_loss['ppm_train_loss'] = ppm_list_train_loss
    ppm_result_loss['ppm_validation_loss'] = ppm_list_validation_loss
    int_result_loss = dict()
    int_result_loss['int_train_loss'] = int_list_train_loss
    int_result_loss['int_validation_loss'] = int_list_validation_loss
    
    return ppm_result_total, int_result_total, ppm_result_loss, int_result_loss
```

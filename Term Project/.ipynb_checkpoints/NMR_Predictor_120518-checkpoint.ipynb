# 1. Pre-Processing

```python
def get_one_inchi(dataset):
    for i in range(len(dataset)):
        partial_list = dataset[i].split('\n')
        dataset[i] = partial_list[0]
    
    return dataset

def atom_feature(atom):
    return np.array(one_of_k_encoding_unk(atom.GetSymbol(),
                                      ['C', 'N', 'O', 'S', 'F', 'H', 'Si', 'P', 'Cl', 'Br',
                                       'Li', 'Na', 'K', 'Mg', 'Ca', 'Fe', 'As', 'Al', 'I', 'B',
                                       'V', 'Tl', 'Sb', 'Sn', 'Ag', 'Pd', 'Co', 'Se', 'Ti', 'Zn',
                                       'Ge', 'Cu', 'Au', 'Ni', 'Cd', 'Mn', 'Cr', 'Pt', 'Hg', 'Pb']) +
                    one_of_k_encoding(atom.GetDegree(), [0, 1, 2, 3, 4, 5]) +
                    one_of_k_encoding_unk(atom.GetTotalNumHs(), [0, 1, 2, 3, 4]) +
                    one_of_k_encoding_unk(atom.GetImplicitValence(), [0, 1, 2, 3, 4, 5]) +
                    [atom.GetIsAromatic()])    # (40, 6, 5, 6, 1)

def one_of_k_encoding(x, allowable_set):
    if x not in allowable_set:
        raise Exception("input {0} not in allowable set{1}:".format(x, allowable_set))
    #print list((map(lambda s: x == s, allowable_set)))
    return list(map(lambda s: x == s, allowable_set))

def one_of_k_encoding_unk(x, allowable_set):
    """Maps inputs not in the allowable set to the last element."""
    if x not in allowable_set:
        x = allowable_set[-1]
    return list(map(lambda s: x == s, allowable_set))

def normalization(dataset, maximum, minimum):
    new_dataset = []
    for i in range(len(dataset)):
        new_dataset.append((dataset[i] - minimum)/(maximum - minimum))
        
    return new_dataset    

def maxNumAtoms(InChi_list):
    NumAtoms = []
    for i in InChi_list:
        iMol = Chem.inchi.MolFromInchi(i)
        iMol = Chem.rdmolops.AddHs(iMol)
        idx = []
        for atom in iMol.GetAtoms():
            idx.append(atom.GetIdx())
        NumAtoms.append(max(idx))
    
    maxNumAtoms = max(NumAtoms)
    
    return maxNumAtoms

def convert_to_graph(InChi_list):
    adj = []
    adj_norm = []
    features = []
    maxNumAtoms = 208
    for i in InChi_list:
        #partial charge
        pc = []
        #Mol
        iMol = Chem.inchi.MolFromInchi(i)
        iMol = Chem.rdmolops.AddHs(iMol)
        #Adj
        iAdjTmp = Chem.rdmolops.GetAdjacencyMatrix(iMol)
        #iMol = mol_with_atom_index(iMol)
        AllChem.ComputeGasteigerCharges(iMol)
        # Feature
        if(iAdjTmp.shape[0] <= maxNumAtoms):
            # Feature-preprocessing
            iFeature = np.zeros((maxNumAtoms, 58))
            iFeatureTmp = []
            for atom in iMol.GetAtoms():
                arr = atom_feature(atom)
                iFeatureTmp.append(arr)
                pc.append(float(atom.GetProp("_GasteigerCharge")))
            max_pc = max(pc)
            min_pc = min(pc)
            pc = normalization(pc, max_pc, min_pc)
            while len(pc) != len(iFeature):
                pc.append(0)
            iFeature[0:len(iFeatureTmp), 0:58] = iFeatureTmp ### 0 padding for feature-set
            for i in range(len(iFeature)):
                iFeature[i][5] = pc[i]
            features.append(iFeature)

            # Adj-preprocessing
            iAdj = np.zeros((maxNumAtoms, maxNumAtoms))
            iAdj[0:len(iFeatureTmp), 0:len(iFeatureTmp)] = iAdjTmp + np.eye(len(iFeatureTmp))
            adj.append(np.asarray(iAdj))
    features = np.asarray(features)

    return features, adj  
```

```python
class GCNDataset(Dataset):
    def __init__(self, list_feature, lise_adj, list_ppm, list_intensity):
        self.list_feature = list_feature
        self.lise_adj = lise_adj
        self.list_ppm = list_ppm
        self.list_intensity = list_intensity
        #self.transform = transforms.Compose([transforms.ToTensor()])  # you can add to the list all the transformations you need. 

    def __len__(self):
        return len(self.list_feature)
            
    def __getitem__(self, index):
        feature = self.list_feature[index]
        adj = self.lise_adj[index]
        ppm = self.list_ppm[index]
        intensity = self.list_intensity[index]
        
        return feature, adj, ppm, intensity
        #return self.transform(self.list_feature[index]), self.transform(self.lise_adj[index]), self.transform(self.list_ppm[index]), self.transform(self.list_intensity[index])

def extract_data(data_list):
    
    # data list = [[Hz list], [ppm list], [intensity list]]
    # if number of ppm/intensity peaks more than 7, extract in order of decreasing from biggest intensity
    intensity_list = []
    _int = []
    _ppm = []

    for i in range(len(data_list)):
        intensity_list.append((i, data_list[i][2]))
    intensity_list.sort(key = lambda x: x[1])
    intensity_list.reverse()
    for i in range(len(data_list)):
        _ppm.append(float(data_list[intensity_list[i][0]][1]))
        _int.append(float(intensity_list[i][1]))
    while len(_ppm) < 150:
        _ppm.append(0)
        _int.append(0)
        
    return _ppm, _int

def normalize_all_data(dataset):
    #input is list of list, find the absolute max and min then, normaize all
    #finding absolute max and min
    abs_max = 0
    abs_min = 0
    for i in range(len(dataset)):
        local_max = max(dataset[i])
        local_min = min(dataset[i])
        if abs_max < local_max:
            abs_max = local_max
        if abs_min > local_min:
            abs_min = local_min
    #print(abs_max, abs_min)
        
    #normalize all with abs_max and abs_min
    new_dataset = []
    for i in range(len(dataset)):
        new_dataset.append(normalization(dataset[i], abs_max, abs_min))
        
    return new_dataset, abs_max, abs_min

def Data(InChi_list, data_list, num_mol):
    feature_total, adj_total = convert_to_graph(InChi_list)
    # intensity_list is in order of decreasing from biggenst intensity value
    #ppm_list is corresponding ppm at that intensity
    ppm_list = []
    intensity_list = []
    for i in range(len(data_list)):
        _ppm, _int = extract_data(data_list[i])
        ppm_list.append((_ppm))
        intensity_list.append((_int))
    
    #print(type(feature_total[0]))
    ppm_dataset, ppm_abs_max, ppm_abs_min = normalize_all_data(ppm_list)
    ppm_total = np.asarray(ppm_dataset)
    ppm_extreme_values = (ppm_abs_max, ppm_abs_min)
    intensity_dataset, intensity_abs_max, intensity_abs_min = normalize_all_data(intensity_list)
    intensity_total = np.asarray(intensity_dataset)
    intensity_extreme_values = (intensity_abs_max, intensity_abs_min)
    #print(type(ppm_total), type(intensity_total))
    #print(ppm_total.shape, intensity_total.shape)

    num_train = int(num_mol*0.6)
    num_validation = int(num_mol*0.2)
    num_test = int(num_mol*0.2)
    
    feature_train = feature_total[0:num_train]
    adj_train = adj_total[0:num_train]
    ppm_train = ppm_total[0:num_train]
    intensity_train = intensity_total[0:num_train]
    
    feature_validation = feature_total[num_train:(num_train+num_validation)]
    adj_validation = adj_total[num_train:(num_train+num_validation)]
    ppm_validation = ppm_total[num_train:(num_train+num_validation)]
    intensity_validation = intensity_total[num_train:(num_train+num_validation)]
    
    feature_test = feature_total[(num_train+num_validation):]
    adj_test = adj_total[(num_train+num_validation):]
    ppm_test = ppm_total[(num_train+num_validation):]
    intensity_test = intensity_total[(num_train+num_validation):]

    train_set = GCNDataset(feature_train, adj_train, ppm_train, intensity_train)
    #print(len(train_set))
    #print(train_set[0][0], train_set[0][1], train_set[0][2], train_set[0][3])
    #print(train_set[0][0].shape, train_set[0][1].shape, len(train_set[0][2]), len(train_set[0][3]))
    validation_set = GCNDataset(feature_validation, adj_validation, ppm_validation, intensity_validation)
    test_set = GCNDataset(feature_test, adj_test, ppm_test, intensity_test)
    partition = {'train' : train_set, 'val' : validation_set, 'test' : test_set}
    
    return partition, ppm_extreme_values, intensity_extreme_values
```

```python
#inchi, mainPeaks, name, peaks, solvent

# deleting NAN data - column is list of column names that we want to delete NaN
def delete_NaN(dataframe):
    NaN_list = []
    column_list = list(dataframe.columns.values)
    #print(column_list)
    for i in range(len(dataframe)):
        if type(dataframe[column_list[0]][i]) != str:
            NaN_list.append(i)
    #print(type(dataframe[column_list[0]][14227]))
    #print(NaN_list)
    
    for i in range(len(NaN_list)):
        dataframe = dataframe.drop([NaN_list[i]], axis = 0)
    dataframe = dataframe.reset_index()
    
    return dataframe

def extract_column(dataframe, column):
    data = []
    for i in range(len(column)):
        if i == 0:
            data.append(get_one_inchi(list(dataframe[column[i]])))
        else:
            data.append(list(dataframe[column[i]]))
    
    return data
    
def delete_overlap(dataset):
    InChi_list = dataset[0]
    peak_list = dataset[1]
    assign = dict()
    for i in range(len(InChi_list)):
        assign[InChi_list[i]] = peak_list[i]
    
    new_dataset = []
    new_InChi_list = list(assign.keys())
    new_peak_list = list(assign.values())
    
    new_dataset.append(new_InChi_list)
    new_dataset.append(new_peak_list)
    
    return new_dataset
```

# 2. Model Construction

```python
class SkipConnection(nn.Module):
    
    def __init__(self, in_dim, out_dim):
        super(SkipConnection, self).__init__()
        
        self.in_dim = in_dim
        self.out_dim = out_dim
        
        self.linear = nn.Linear(in_dim, out_dim, bias=False)
        
    def forward(self, in_x, out_x):
        if (self.in_dim != self.out_dim):
            in_x = self.linear(in_x)
        out = in_x + out_x
        return out

class GatedSkipConnection(nn.Module):
    
    def __init__(self, in_dim, out_dim):
        super(GatedSkipConnection, self).__init__()
        
        self.in_dim = in_dim
        self.out_dim = out_dim
        
        self.linear = nn.Linear(in_dim, out_dim, bias=False)
        self.linear_coef_in = nn.Linear(out_dim, out_dim)
        self.linear_coef_out = nn.Linear(out_dim, out_dim)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, in_x, out_x):
        if (self.in_dim != self.out_dim):
            in_x = self.linear(in_x)
        z = self.gate_coefficient(in_x, out_x)
        out = torch.mul(z, out_x) + torch.mul(1.0-z, in_x)
        return out
            
    def gate_coefficient(self, in_x, out_x):
        x1 = self.linear_coef_in(in_x)
        x2 = self.linear_coef_out(out_x)
        return self.sigmoid(x1+x2)
```

```python
class Attention(nn.Module):
    
    def __init__(self, in_dim, output_dim, num_head):
        super(Attention, self).__init__()
        
        self.num_head = num_head
        self.atn_dim = output_dim // num_head
        
        self.linears = nn.ModuleList()
        self.corelations = nn.ParameterList()
        for i in range(self.num_head):
            self.linears.append(nn.Linear(in_dim, self.atn_dim))
            corelation = torch.FloatTensor(self.atn_dim, self.atn_dim)
            nn.init.xavier_uniform_(corelation)
            self.corelations.append(nn.Parameter(corelation))
            
        self.tanh = nn.Tanh()
        
    def forward(self, x, adj):
        heads = list()
        for i in range(self.num_head):
            x_transformed = self.linears[i](x)
            alpha = self.attention_matrix(x_transformed, self.corelations[i], adj)
            x_head = torch.matmul(alpha, x_transformed)
            heads.append(x_head)
        output = torch.cat(heads, dim=2)
        return output
            
    def attention_matrix(self, x_transformed, corelation, adj):
        x = torch.einsum('akj,ij->aki', (x_transformed, corelation))
        alpha = torch.matmul(x, torch.transpose(x_transformed, 1, 2))
        alpha = torch.mul(alpha, adj)
        alpha = self.tanh(alpha)
        return alpha
```

```python
class GCNLayer(nn.Module):
    
    def __init__(self, in_dim, out_dim, n_atom, act=None, bn=False, atn=False, num_head=1, dropout=0):
        super(GCNLayer, self).__init__()
        
        self.use_bn = bn
        self.use_atn = atn
        self.linear = nn.Linear(in_dim, out_dim)
        nn.init.xavier_uniform_(self.linear.weight)
        self.bn = nn.BatchNorm1d(n_atom)
        self.attention = Attention(out_dim, out_dim, num_head)
        self.activation = act
        self.dropout_rate = dropout
        self.dropout = nn.Dropout2d(self.dropout_rate)
        
    def forward(self, x, adj):
        out = self.linear(x)
        if self.use_atn:
            out = self.attention(out, adj)
        else:
            out = torch.matmul(adj, out)
        if self.use_bn:
            out = self.bn(out)
        if self.activation != None:
            out = self.activation(out)
        if self.dropout_rate > 0:
            out = self.dropout(out)
        return out, adj
```

```python
class GCNBlock(nn.Module):
    
    def __init__(self, n_layer, in_dim, hidden_dim, out_dim, n_atom, bn=True, atn=True, num_head=1, sc='gsc', dropout=0):
        super(GCNBlock, self).__init__()
        
        self.layers = nn.ModuleList()
        for i in range(n_layer):
            self.layers.append(GCNLayer(in_dim if i==0 else hidden_dim,
                                        out_dim if i==n_layer-1 else hidden_dim,
                                        n_atom,
                                        nn.ReLU() if i!=n_layer-1 else None,
                                        bn,
                                        atn,
                                        num_head,
                                        dropout))
        self.relu = nn.ReLU()
        if sc=='gsc':
            self.sc = GatedSkipConnection(in_dim, out_dim)
        elif sc=='sc':
            self.sc = SkipConnection(in_dim, out_dim)
        elif sc=='no':
            self.sc = None
        else:
            assert False, "Wrong sc type."
        
    def forward(self, x, adj):
        residual = x
        for i, layer in enumerate(self.layers):
            out, adj = layer((x if i==0 else out), adj)
        if self.sc != None:
            out = self.sc(residual, out)
        out = self.relu(out)
        return out, adj
```

```python
class ReadOut(nn.Module):
    
    def __init__(self, in_dim, out_dim, act=None):
        super(ReadOut, self).__init__()
        
        self.in_dim = in_dim
        self.out_dim= out_dim
        
        self.linear = nn.Linear(self.in_dim, 
                                self.out_dim)
        nn.init.xavier_uniform_(self.linear.weight)
        self.activation = act

    def forward(self, x):
        out = self.linear(x)
        out = torch.sum(out, 1)
        if self.activation != None:
            out = self.activation(out)
        return out
```

```python
class Predictor(nn.Module):
    
    def __init__(self, in_dim, out_dim, act=None):
        super(Predictor, self).__init__()
        
        self.in_dim = in_dim
        self.out_dim = out_dim
        
        self.linear = nn.Linear(self.in_dim,
                                self.out_dim)
        nn.init.xavier_uniform_(self.linear.weight)
        self.activation = act
        
    def forward(self, x):
        out = self.linear(x)
        if self.activation != None:
            out = self.activation(out)
        return out
```

```python
class GCNNet(nn.Module):
    
    def __init__(self, args):
        super(GCNNet, self).__init__()
        
        self.blocks = nn.ModuleList()
        for i in range(args.n_block):
            self.blocks.append(GCNBlock(args.n_layer,
                                        args.in_dim if i==0 else args.hidden_dim,
                                        args.hidden_dim,
                                        args.hidden_dim,
                                        args.n_atom,
                                        args.bn,
                                        args.atn,
                                        args.num_head,
                                        args.sc,
                                        args.dropout))
        self.readout = ReadOut(args.hidden_dim, 
                               args.readout_dim,
                               act=nn.ReLU())
        self.shift_pred1 = Predictor(args.readout_dim,
                               args.shift_pred_dim1,
                               act=nn.ReLU())
        self.shift_pred2 = Predictor(args.shift_pred_dim1,
                               args.shift_pred_dim2,
                               act=nn.ReLU())
        self.shift_pred3 = Predictor(args.shift_pred_dim2,
                               args.out_dim)
        self.intensity_pred1 = Predictor(args.readout_dim,
                                         args.intensity_pred_dim1,
                                         act=nn.ReLU())
        self.intensity_pred2 = Predictor(args.intensity_pred_dim1,
                                         args.intensity_pred_dim2,
                                         act=nn.ReLU())
        self.intensity_pred3 = Predictor(args.intensity_pred_dim2,
                                         args.out_dim)
        
    def forward(self, x, adj):
        for i, block in enumerate(self.blocks):
            out, adj = block((x if i==0 else out), adj)
        out = self.readout(out)
        shift_out = self.shift_pred1(out)
        shift_out = self.shift_pred2(shift_out)
        shift_out = self.shift_pred3(shift_out)
        intensity_out = self.intensity_pred1(out)
        intensity_out = self.intensity_pred2(intensity_out)
        intensity_out = self.intensity_pred3(intensity_out)
        return shift_out, intensity_out
```

# 3. Train, Validate, and Test

```python
def train(model, device, optimizer, criterion, data_train, bar, args):
    epoch_train_loss = 0
    for i, batch in enumerate(data_train):
        # [batch_size, max_atom, num_feature], [100, 250, 59]
        list_feature = torch.tensor(batch[0]).to(device).float()
        list_adj = torch.tensor(batch[1]).to(device).float()
        # [batch_size, max_peak]
        # [150, 100]
        #print(len(batch[2]), len(batch[2][0]))
        list_shift = torch.tensor(batch[2]).to(device).float()
        list_intensity = torch.tensor(batch[3]).to(device).float()
        list_shift = list_shift.view(-1,1)
        list_intensity = list_intensity.view(-1,1)        
        
        model.train()
        optimizer.zero_grad()
        
        list_pred_shift, list_pred_intensity = model(list_feature, list_adj)
        list_pred_shift.require_grad = False
        list_pred_intensity.require_grad = False
        
        list_pred_shift = list_pred_shift.view(-1, 1)
        list_pred_intensity = list_pred_intensity.view(-1, 1)
        train_shift_loss = criterion(list_pred_shift, list_shift)
        train_intensity_loss = criterion(list_pred_intensity, list_intensity)
        
        train_loss = train_shift_loss + train_intensity_loss
        epoch_train_loss += train_loss.item()
        
        train_loss.backward()
        optimizer.step()
       
        bar.update(len(list_feature))

    epoch_train_loss /= len(data_train)
    
    return model, epoch_train_loss
```

```python
def validate(model, device, optimizer, criterion, data_val, bar, args):
    epoch_val_loss = 0
    for i, batch in enumerate(data_val):
        list_feature = torch.tensor(batch[0]).to(device).float()
        list_adj = torch.tensor(batch[1]).to(device).float()
        list_shift = torch.tensor(batch[2]).to(device).float()
        list_intensity = torch.tensor(batch[3]).to(device).float()
        list_shift = list_shift.view(-1,1)
        list_intensity = list_intensity.view(-1,1)        
        
        model.eval()
        
        list_pred_shift, list_pred_intensity = model(list_feature, list_adj)
        list_pred_shift.require_grad = False
        list_pred_intensity.require_grad = False
        
        list_pred_shift = list_pred_shift.view(-1, 1)
        list_pred_intensity = list_pred_intensity.view(-1, 1)
        val_shift_loss = criterion(list_pred_shift, list_shift)
        val_intensity_loss = criterion(list_pred_intensity, list_intensity)
        
        val_loss = val_shift_loss + val_intensity_loss
        epoch_val_loss += val_loss.item()
       
        bar.update(len(list_feature))

    epoch_val_loss /= len(data_val)
    
    return model, epoch_val_loss
```

```python
def test(model, device, data_test, args):
    model.eval()
    with torch.no_grad():
        shift_total = list()
        pred_shift_total = list()
        intensity_total = list()
        pred_intensity_total = list()
        for i, batch in enumerate(data_test):
            list_feature = torch.tensor(batch[0]).to(device).float()
            list_adj = torch.tensor(batch[1]).to(device).float()
            list_shift = torch.tensor(batch[2]).to(device).float()
            list_intensity = torch.tensor(batch[3]).to(device).float()
            #shape [bs x max atom x num feature], [bs x max atom x max atom]
            #shape [bs x num peaks], [bs x num peaks]
            #print(batch[0].shape, batch[1].shape, batch[2].shape, batch[3].shape)
            list_shift = list_shift.view(-1,1)
            list_intensity = list_intensity.view(-1,1)
            shift_total.append(list_shift.tolist())
            intensity_total.append(list_intensity.tolist())

            list_pred_shift, list_pred_intensity = model(list_feature, list_adj)
            #shape [bs x num peaks], [bs x num peaks]
            #print(list_pred_shift.shape, list_pred_intensity.shape)
            pred_shift_total.append(list_pred_shift.view(-1, 1).tolist())
            pred_intensity_total.append(list_pred_intensity.view(-1, 1).tolist())
        
        mae_shift = 0
        mae_intensity = 0
        #print(shift_total[0])
        #print(pred_shift_total[0])
        for i in range(len(shift_total)):
            mae_shift += mean_absolute_error(shift_total[i], pred_shift_total[i])/len(shift_total[i])
            mae_intensity += mean_absolute_error(intensity_total[i], pred_intensity_total[i])/len(intensity_total[i])
        #mae_shift /= len(shift_total)
        #mae_intensity /= len(intensity_total)
        print(mae_shift, mae_intensity)
        
    return mae_shift, mae_intensity, shift_total, pred_shift_total, intensity_total, pred_intensity_total
```

```python
def experiment(dict_partition, device, bar, args):
    time_start = time.time()
    
    model = GCNNet(args)
    model.to(device)
    
    if args.optim == 'Adam':
        optimizer = optim.Adam(model.parameters(),
                               lr=args.lr,
                               weight_decay=args.l2_coef)
    elif args.optim == 'RMSprop':
        optimizer = optim.RMSprop(model.parameters(),
                               lr=args.lr,
                               weight_decay=args.l2_coef)
    elif args.optim == 'SGD':
        optimizer = optim.SGD(model.parameters(),
                               lr=args.lr,
                               weight_decay=args.l2_coef)
    else:
        assert False, 'Undefined Optimizer Type'
        
    criterion = nn.MSELoss()
    scheduler = optim.lr_scheduler.StepLR(optimizer,
                                          step_size=args.step_size,
                                          gamma=args.gamma)
    
    list_train_loss = list()
    list_val_loss = list()

    data_train = DataLoader(dict_partition['train'], 
                            batch_size=args.batch_size,
                            shuffle=True)

    data_val = DataLoader(dict_partition['val'],
                          batch_size=args.batch_size,
                          shuffle=False)
    
    for epoch in range(args.epoch):
        scheduler.step()
        model, train_loss = train(model, device, optimizer, criterion, data_train, bar, args)
        list_train_loss.append(train_loss)
        
        model, val_loss = validate(model, device, optimizer, criterion, data_val, bar, args)
        list_val_loss.append(val_loss)
        
    data_test = DataLoader(dict_partition['test'],
                           batch_size=args.batch_size,
                           shuffle=args.shuffle)

    mae_shift, mae_intensity, shift_total, pred_shift_total, intensity_total, pred_intensity_total = test(model, device, data_test, args)
    
    time_end = time.time()
    time_required = time_end - time_start
    
    args.list_train_loss = list_train_loss
    args.list_val_loss = list_val_loss
    args.shift_total = shift_total
    args.pred_shift_total = pred_shift_total
    args.intensity_total = intensity_total
    args.pred_intensity_total = pred_intensity_total
    args.mae_shift = mae_shift
    args.mae_intensity = mae_intensity
    args.time_required = time_required
    
    return args
```

```python
#ppm_abs_max, ppm_abs_min
#intensity_abs_max, intensity_abs_min
#args.pred_shift_total, args.pred_intensity_total

def denormalize_all_data(dataset, maximum, minimum):
    #input data is list of list form
    new_dataset = []
    for i in range(len(dataset)):
        sub_dataset = []
        for j in range(len(dataset[i])):
            denormalized_value = dataset[i][j][0]*(maximum - minimum) + minimum
            sub_dataset.append(denormalized_value)
        new_dataset.append(sub_dataset)
    
    return new_dataset
#print(args.pred_shift_total, '\n')
#print(denormalize_all_data(args.pred_shift_total, ppm_abs_max, ppm_abs_min))
```

```python
def plot_performance(df_result, var1, var2):
    fig, ax = plt.subplots(1, 2)
    fig.set_size_inches(10, 5)
    
    df_mae = df_result.pivot(var1, var2, 'mae_shift')
    df_std = df_result.pivot(var1, var2, 'mae_intensity')
    df_mae = df_mae[df_mae.columns].astype(float)
    df_std = df_std[df_std.columns].astype(float)
    
    hm_mae = sns.heatmap(df_mae, ax=ax[0], annot=True, fmt='f', linewidths=.5, cmap='YlGnBu')
    hm_std = sns.heatmap(df_std, ax=ax[1], annot=True, fmt='f', linewidths=.5, cmap='YlGnBu')
    
    fig.suptitle('Performance depends on ' + var1 + ' vs ' + var2)
    hm_mae.set_title('Shift MAE depends on ' + var1 + ' vs ' + var2)
    hm_std.set_title('Intensity MAE depends on ' + var1 + ' vs ' + var2)    
```

```python
def plot_performance_bar(df_result, var1, var2, ylim):
    fig, ax = plt.subplots(1, 2)
    fig.set_size_inches(10, 5)
    
    sns.set_style("darkgrid", {"axes.facecolor": ".9"})
    bar_mae = sns.barplot(x=var1, y='mae_shift', hue=var2, data=df_result, ax=ax[0])
    bar_std = sns.barplot(x=var1, y='mae_intensity', hue=var2, data=df_result, ax=ax[1])
    
    bar_mae.set_title('Shift MAE depends on ' + var1 + ' vs ' + var2)
    bar_std.set_title('Std depends on ' + var1 + ' vs ' + var2)
    sns.plt.ylim(0.0, ylim)
    fig.suptitle('Intensity MAE depends on ' + var1 + ' vs ' + var2)
```

```python
def plot_loss(df_result, var1, var2, ylim):
    def plot(x, ylim=1.0, **kwargs):
        plt.plot(x[0], **kwargs)
        plt.ylim(0.0, ylim)
    
    sns.set_style("darkgrid", {"axes.facecolor": ".9"})
    g = sns.FacetGrid(df_result, row=var1, col=var2, margin_titles=True)
    g.map(plot, 'list_train_loss', ylim=ylim, label='Train Loss')
    g.map(plot, 'list_val_loss', ylim=ylim, color='r', label='Validation Loss')
    g.fig.suptitle('Loss vs Epochs depends on ' + var1 + ' vs ' + var2, size=16)
    g.fig.subplots_adjust(top=.9)
    plt.legend()
    plt.show()
```

```python
def plot_distribution(df_result, var1, var2, ylim):
    def scatter(x, y, **kwargs):
        plt.scatter(x[0][0], y[0][0], alpha=0.3, s=2)
    def identity(x, y, **kwargs):
        plt.plot(x[0][0], x[0][0], alpha=0.4, color='black')
        plt.ylim(0.0, ylim)
    
    sns.set_style("darkgrid", {"axes.facecolor": ".9"})
    g = sns.FacetGrid(df_result, row=var1, col=var2, margin_titles=True)
    g.map(scatter, 'shift_total', 'pred_shift_total')
    g.map(identity, 'shift_total', 'shift_total')
    g.fig.suptitle('Truth Distribution depends on ' + var1 + ' vs ' + var2, size=16)
    g.fig.subplots_adjust(top=.9)
    plt.show()
```

```python
%%time
import numpy as np
from utils import read_data
from rdkit import Chem, DataStructs
from rdkit.Chem import AllChem
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score, accuracy_score
```

```python
tox_type = 'sr-p53'
smiles, label = read_data(tox_type)
```

# Getting molecular fingerprints of each molecules

```python
mols = np.array([Chem.MolFromSmiles(smile) for smile in smiles])
fps = []
labels = []
for i in range(len(mols)):
    try:
        fp = AllChem.GetMorganFingerprintAsBitVect(mols[i],2)
        arr = np.zeros((1,))
        DataStructs.ConvertToNumpyArray(fp, arr)
        fps.append(arr)
        labels.append(label[i])
    except:
        print("This molecule cannot be converted to the molecular fingerprint by RDKit.")
fps = np.asarray(fps)
labels = np.asarray(labels)
```

# Splitting the dataset to train set and test set

```python
fps.shape, labels.shape
```

```python
num = fps.shape[0]
num_train = int(num * 0.8)
num_test = num - num_train

num, num_train, num_test
```

```python
fps_train = fps[:num_train]
labels_train = labels[:num_train]
fps_test = fps[num_train:]
labels_test = labels[num_train:]

fps_train.shape, labels_train.shape
```

# Training a SVM classification model

```python
_gamma = 5.0
clf = SVC(kernel='poly', gamma=_gamma)
clf.fit(fps_train, labels_train)
```

# Validating the train model with the test set

```python
labels_pred = clf.predict(fps_test)
labels_score = clf.decision_function(fps_test)
auc = accuracy_score(labels_test, labels_pred)
ras = roc_auc_score(labels_test, labels_score)
auc, ras
```
